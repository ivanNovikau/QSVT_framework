```python
alpha = 0.1
n = 10
sep_angles = [alpha/2.**xx for xx in range(n)]
print(sep_angles)

print(np.sum(sep_angles))
print(2*alpha * (1 - 0.5**n))

res_angles = np.zeros(2**n)
for ii in range(2**n):
    bit_array = mix.find_bit_array_of_int(ii, n)
#     print(bit_array)
    res_angles[ii] = np.sum([sep_angles[iqq] * bit_array[iqq] for iqq in range(n)])
print(res_angles)
```

```python
# option 1
res_x = np.sin(res_angles)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(res_angles, res_x, color='b', marker = "o", linewidth = 2, linestyle='-', label = "option 1")
plt.xlabel('angles')
plt.ylabel("res. x-coordinate")
ax.legend()
plt.grid(True)
plt.show()
```
